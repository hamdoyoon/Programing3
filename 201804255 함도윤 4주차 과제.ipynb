# 과제 1번  

movies 데이터 줄이기

```python
import pandas as pd
import numpy as np
```

```python
movie = pd.read_csv('data/movie.csv')
```

```python
movie1 = movie.iloc[:, :11].fillna(0)
```

```python
데이터에 대한 dtype을 확인하기위해 info를 실행시켜주었습니다.
movie1.info()
```

```python
# 수치형 데이터의 max 값을 확인하기 위해 실행해주었습니다.
movie1.describe().T
```

```python
# object형 경우 category로 만들어 줄 필요성이 있는지 먼저 검색해보았습니다.
# 개별적으로 많이 나눠져있는 경우 굳이 category로 묶어주는 이유가 없기 때문입니다.
movie1.color.value_counts()
```

```python
#memory_usage 메서드를 사용하면, 메모리 사용량을 컬럼별로 알 수 있습니다.
#col2.memory_usage()
#일반적으로 object 타임은 정확한 계산이 되지 않으므로, deep=True로 바꿔줍니다.
original_mem = movie1.memory_usage(deep=True)
original_mem
```

```python
movie1['num_critic_for_reviews'] = movie1['num_critic_for_reviews'].astype(np.float16)   
movie1.memory_usage(deep=True)
```

```python
#director_name의 경우 category로 변경해주어도 용량의 변함이 거의 없었습니다.
movie1['duration'] = movie1['duration'].astype(np.float16)
movie1['director_facebook_likes'] = movie1['director_facebook_likes'].astype(np.float16)
movie1['actor_3_facebook_likes'] = movie1['actor_3_facebook_likes'].astype(np.float16)
movie1['gross'] = movie1['gross'].astype(np.float16)
movie1['color'] = movie1['color'].astype('category')
movie1['genres'] = movie1['genres'].astype('category')
movie1['actor_1_name'] = movie1['actor_1_name'].astype('category')
movie1['actor_2_name'] = movie1['actor_2_name'].astype('category')

new_mem = movie1.memory_usage(deep=True)
new_mem
```

```python
#이전 메모리와 새로운 메모리의 절약된 비율입니다.
new_mem / original_mem
```

# 과제 2번  

UGDS의 도시별 최대값 구하기

```python
college = pd.read_csv('data/college.csv')
college.head()
```

```python
#도시별 UGDS의 최댓값을 구하는데
#필요한 열을 추출하고 도시별로 정렬한 다음 UGDS의 최대값을 찾았습니다.
#내림차순으로 ascending=False를 해주었습니다.
(college
 [['INSTNM', 'UGDS', 'CITY']]
 .sort_values(['CITY','UGDS'], ascending=False)
)
```

# 과제 3번  

주별, RELAFFIL 별 UGDS의 최대값을 구하라.  

INSTNM, STABBR, RELAFFIL, UGDS 이런식으로 컬럼 만들어서 가져오기

```python
#먼저 STABBR(주별)을 먼저 정렬을 해주고 그 다음 RELAFFIL(종교)를 찾고
#그 종교별 UGDS의 최댓값을 찾았습니다.
(college
 [['INSTNM','STABBR', 'RELAFFIL','UGDS']]
 .sort_values(['STABBR','RELAFFIL','UGDS'], ascending=False)
)
```
